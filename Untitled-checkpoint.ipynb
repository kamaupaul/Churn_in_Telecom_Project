### 1. Business Understanding

The primary objective of any business is to reduce customer churn. By identifying potential churners in advance, the company can take precaution measures to retain these customers.

Such measures may include: 

    - Improved customer service.

    - Addressing the issues that may lead to churn.

    - Making targeted offers and,

    - Targeted market campaigns among others.
    
Churn can have great significant financial impact on the business as high churn leads to loss of recurring revenue, damage of brand's reputation among many other effects.

This project aims to predict customer churn by developing an algorithm to predict the churn rate based on customer usage pattern.

##### Checking for Outliers.

```python
# plotting a boxplot of each column to view outliers.

numeric_columns = df.select_dtypes(include='number')
plt.figure(figsize=(20, 10))
plt.boxplot(numeric_columns.values)
plt.title("Box plot of all numeric columns in the dataset")
plt.xticks(range(1, len(numeric_columns.columns) + 1), numeric_columns.columns,rotation=90)
plt.show()
```

Outliers can affect the predictive power of our models,therefore removing them will increase our model performance.

```python
# Removing outliers 
for col in numeric_columns: 
    q1 = df[col].quantile(0.20)
    q3 = df[col].quantile(0.80)
    iqr = q3-q1
    range_low  = q1-1.5*iqr
    range_high = q3+1.5*iqr
    # Assigning the filtered dataset into data
    data = df.loc[(df[col] > range_low) & (df[col] < range_high)]

data.shape
```

###  3. Univariate analysis

Exploring some of the variables in the data set looking for patterns of response to the variable.

```python
# plot a bar graph to show the distribution of churn
data['churn'].value_counts().plot(kind='bar')
plt.xlabel('churn')
plt.ylabel('count')
plt.title('Distribution of Churn')
data['churn'].value_counts(normalize=True)
```

The bar graph above shows the distribution of churn where we see 14% of the customers churned.

```python
# compare the churn rates between customers with and without a voice mail plan.
pivot_df = data.pivot_table(values='churn', index='voice mail plan', aggfunc='mean')
pivot_df.plot(kind='bar')
plt.ylabel('Churn Rate')
plt.xlabel('voice mail plan')
plt.title('Churn Rate by voice mail plan')
plt.show()
```

From the plot above we see that the customers who had no voice mail plan had the highest churn rate.

```python
# Checking the distribution of columns.
plt.figure(figsize=(12, 8))
for i, col in enumerate(numeric_columns, 1):
    plt.subplot(4, 4, i)
    sns.histplot(df[col], kde=True)
    plt.xlabel(col)
plt.tight_layout()
plt.show();
```

As seen from the distributions above most of the numeric features are normally distributed and the features will be good for modelling.

```python
# compare the churn rates between customers who had customer service calls
pivot_df = data.pivot_table(values='churn', index='customer service calls', aggfunc='mean')
pivot_df.plot(kind='bar')
plt.ylabel('Churn Rate')
plt.xlabel('customer service calls')
plt.title('Churn Rate by customer service calls')
plt.show()
```

From the plot above, a customer who had a customer service call are most likely to churn.
This implies the costomers who did not have a customer service call are most likely satisfied from the services offered by the company.

### 3.2 Bivariate analysis.

In this analysis, we will be examining the bivariate relationships between the features we have

```python
# function for plotting the scatter plots
def scatterplot(df, x_col, y_col, hue_col):
    """Plots a scatterplot between two variables in a pandas DataFrame"""
    x_data = df[x_col]
    y_data = df[y_col]
    hue= df[hue_col]
    sns.scatterplot(data=df, x=x_data, y=y_data, hue=hue_col)
```

```python
scatterplot(data, 'total day minutes', 'total day calls', 'churn')
```

From the plot above most of the customers who churned are the customers with many total day minutes.

```python
scatterplot(data, 'international plan', 'account length', 'churn')
```
